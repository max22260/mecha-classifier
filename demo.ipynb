```python
# Get filenames in list
from os import listdir
from os.path import isfile, join

mypath = "./output/"

file_names = [f for f in listdir(mypath) if isfile(join(mypath, f))]

print(str(len(file_names)) + ' images loaded')
```

```python
import cv2
import numpy as np
import sys
import os
import shutil

# Extract 1000 for our training data and 500 for our validation set
# Takes about ~20 seconds to run
cooler_count = 0
HBP_count = 0
turbo_count = 0
training_size = 400
test_size = 200
training_images = []
training_labels = []
test_images = []
test_labels = []
size = 150
cooler_dir_train = "./datasets/train/cooler/"
HBP_dir_train = "./datasets/train/HBP/"
turbo_dir_train = "./datasets/train/Turbo/"
cooler_dir_val = "./datasets/validation/cooler/"
HBP_dir_val = "./datasets/validation/HBP/"
turbo_dir_val = "./datasets/validation/turbo/"


def make_dir(directory):
        if os.path.exists(directory):
            shutil.rmtree(directory)
        os.makedirs(directory)

make_dir(cooler_dir_train)
make_dir(HBP_dir_train)
make_dir(turbo_dir_train)
make_dir(cooler_dir_val)
make_dir(HBP_dir_val)
make_dir(turbo_dir_val)


def getZeros(number):
    if(number > 10 and number < 100):
        return "0"
    if(number < 10):
        return "00"
    else:
        return ""

for i, file in enumerate(file_names):
    
    if file_names[i][0] == "C":
        cooler_count += 1
        image = cv2.imread(mypath+file)
        image = cv2.resize(image, (size, size), interpolation = cv2.INTER_AREA)
        if cooler_count <= training_size:
            training_images.append(image)
            training_labels.append(1)
            zeros = getZeros(cooler_count)
            cv2.imwrite(cooler_dir_train + "cooler" + str(zeros) + str(cooler_count) + ".jpg", image)
        if cooler_count > training_size and cooler_count <= training_size+test_size:
            test_images.append(image)
            test_labels.append(1)
            zeros = getZeros(cooler_count-1000)
            cv2.imwrite(cooler_dir_val + "cooler" + str(zeros) + str(cooler_count-1000) + ".jpg", image)
            
    if file_names[i][0] == "H":
        HBP_count += 1
        image = cv2.imread(mypath+file)
        image = cv2.resize(image, (size, size), interpolation = cv2.INTER_AREA)
        if HBP_count <= training_size:
            training_images.append(image)
            training_labels.append(0)
            zeros = getZeros(HBP_count)
            cv2.imwrite(HBP_dir_train + "HBP" + str(zeros) + str(HBP_count) + ".jpg", image)
        if HBP_count > training_size and HBP_count <= training_size+test_size:
            test_images.append(image)
            test_labels.append(0)
            zeros = getZeros(HBP_count-1000)
            cv2.imwrite(HBP_dir_val + "HBP" + str(zeros) + str(HBP_count-1000) + ".jpg", image)
    
    if file_names[i][0] == "T":
        turbo_count += 1
        image = cv2.imread(mypath+file)
        image = cv2.resize(image, (size, size), interpolation = cv2.INTER_AREA)
        if turbo_count <= training_size:
            training_images.append(image)
            training_labels.append(2)
            zeros = getZeros(turbo_count)
            cv2.imwrite(turbo_dir_train + "turbo" + str(zeros) + str(turbo_count) + ".jpg", image)
        if turbo_count > training_size and turbo_count <= training_size+test_size:
            test_images.append(image)
            test_labels.append(2)
            zeros = getZeros(turbo_count-1000)
            cv2.imwrite(turbo_dir_val + "turbo" + str(zeros) + str(turbo_count-1000) + ".jpg", image)
            

    if cooler_count == training_size+test_size and HBP_count == training_size+test_size and turbo_count == training_size+test_size :
        break

print("Training and Test Data Extraction Complete")
```

```python
# Using numpy's savez function to store our loaded data as NPZ files
np.savez('m_training_data.npz', np.array(training_images))
np.savez('m_training_labels.npz', np.array(training_labels))
np.savez('m_test_data.npz', np.array(test_images))
np.savez('m_test_labels.npz', np.array(test_labels))
```

```python
# Loader Function
import numpy as np

def load_data_training_and_test(datasetname):
    
    npzfile = np.load(datasetname + "_training_data.npz")
    train = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_training_labels.npz")
    train_labels = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_test_data.npz")
    test = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_test_labels.npz")
    test_labels = npzfile['arr_0']

    return (train, train_labels), (test, test_labels)
```

```python
from keras.utils import np_utils
(x_train, y_train), (x_test, y_test) = load_data_training_and_test("m")
# Reshaping our label data from (2000,) to (2000,1) and test data from (1000,) to (1000,1)
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)

# Change our image type to float32 data type
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# Normalize our data by changing the range from (0 to 255) to (0 to 1)
x_train /= 255
x_test /= 255

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
num_classes = y_test.shape[1]
print(num_classes)
```

```python
from __future__ import print_function
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
import os

batch_size = 16
epochs = 100

img_rows = x_train[0].shape[0]
img_cols = x_train[1].shape[0]
input_shape = (img_rows, img_cols, 3)

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Dense(num_classes,activation='softmax'))



model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

print(model.summary())
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test, y_test),
          shuffle=True)

model.save("Trained Models/model_V3.h5")

# Evaluate the performance of our trained model
scores = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```
